```python
#подсчет пропусков в датасете
def bar_perc(data, figsize,fontsize, kind = 'bar', 
             color = 'royalblue', colormap = None, 
             x = None, y = None, legend = True,
            grid = True, rot = None, edgecolor = 'black',
            linewidth = 2):
    #buf = data.loc[data.isna() == True].isna().sum()
    buf = (data.loc[data.isna().any(axis = 1)]
           .isna().sum()
          )
    buf.apply(lambda x: x*100/data.shape[0]).plot(
        kind = kind,
        figsize = figsize,
        colormap = colormap,
        x = x,
        y = y,
        rot = rot,
        fontsize = fontsize,
        legend = legend,
        edgecolor = edgecolor,
        linewidth = linewidth,
        color = color,
        grid = grid
    );
```

```python
#pairplot
def sea_pair(data,title, palette = 'Spectral', 
             hue = None,diag_kind = 'hist',
             kind = 'scatter', height = 3,aspect = 4, ):
    sns.pairplot(
        data = data,
        palette = palette,
        diag_kind = diag_kind,
        height = height,
        aspect = aspect,
        hue = hue,
        kind = kind
    ).set(title = title)
```

```python
#corr matrix
def matrix(data, title, figsize, corr_min):
    dfCorr = gold_train.corr()
    filteredDf = dfCorr[((dfCorr >= corr_min) | (dfCorr <= -1*corr_min)) & (dfCorr !=1.000)]
    filteredDf = filteredDf.loc[filteredDf.notna().any(axis = 1)]
    filteredDf = filteredDf.loc[filteredDf.notna().any(axis = 0)]
    plt.figure(figsize = figsize)
    sns.heatmap(filteredDf, annot=True, cmap="RdBu")
    plt.title(title)
    plt.show()
```

```python
#boxplot
def box(data, figsize, x = None, y = None, showfliers = False,
        showmeans = True, hue = None, dodge = False):      
    sns.set(rc={"figure.figsize":figsize})
    sns.boxplot(
        x = x,
        y = y, 
        data = data,
        showfliers = showfliers,
        hue = hue,
        dodge = dodge,
        showmeans = True, # отображение средего для каждого ящика
        meanprops={"marker":"o",
                           "markerfacecolor":"white", 
                           "markeredgecolor":"black",
                          "markersize":"10"}
    );
```

```python
#гистограмма распределений
def hist(data, figsize,title, density = True, bins = 50, 
         alpha = .75, ax = None, legend = False, color = None,
         fontsize = 12, label = None):
    number = 25
    cmap = plt.get_cmap('tab10')
    colors = [cmap(i) for i in np.linspace(0, 1, number)]
    if color is None:
        data.plot(
            kind = 'hist',
            figsize = figsize,
            alpha = alpha,
            color = colors[np.random.randint(25)],
            ax = ax,
            bins = bins,
            legend = legend,
            label = label,
            density = density
        ).set_title(title + "\n", fontsize = fontsize)
    else:
        data.plot(
            kind = 'hist',
            figsize = figsize,
            alpha = alpha,
            color = color,
            ax = ax,
            bins = bins,
            legend = legend,
            label = label,
            density = density
        ).set_title(title + "\n", fontsize = fontsize)
        
```

```python
#функция вычисления эффективности обогащения, где
#C — доля золота в концентрате после флотации/очистки;
#F — доля золота в сырье/концентрате до флотации/очистки;
#T — доля золота в отвальных хвостах после флотации/очистки.
def recovery(c,f,t):
    recovery = (100*c*(f-t))/(f*(c-t))
    return recovery
```

```python
#новая smape
def smape_unweighted(y_true, y_pred):
    error = abs(y_true - y_pred)
    scale = (abs(y_true) + abs(y_pred)) / 2 + 0.000000001
    result = (error / scale).mean()*100
    return result
```

```python
#функция вычисления итоговой sMAPE
# Нужно спрогнозировать сразу две величины:
# эффективность обогащения чернового концентрата rougher.output.recovery;
# эффективность обогащения финального концентрата final.output.recovery.
# Итоговая метрика складывается из двух величин: sMAPE(rough) и sMAPE(final)
def tot_smape(rough,final):
    return(K_ROUGH * rough + K_FINAL * final)
```

Из графика видно, что распределения из двух разных выборок достаточно хорошо совпадают

Применим `t-test`, чтобы понять отличаются ли два распределения статистически значимо.  
*H0* - статистически значимо средние двух ГС не отличаются друг от друга;  

*H1* - распределения средних двух ГС различаются статистически значимо

```python
#p-value
alpha = 0.05

stat, p_value = ttest_ind(
    gold_train['rougher.input.feed_size'],
    gold_test['rougher.input.feed_size']
)
print(f"Значения t-test: разница двух средних ={stat:.4f}, p-value={p_value:.4f}")
if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Итак, **статистически значимо данные не отличаются**, продолжаем работу
